```r
data(mpg, package="ggplot2")
theme_set(theme_bw())  # pre-set the bw theme.
```

```r
mpg %>%
  kable("html") %>%
  as.character() %>%
  display_html()
```

```r
g <- ggplot(data = mpg) + aes(x=cty,y= hwy) +geom_count(color='tomato3') +
  labs(subtitle="mpg: city vs highway mileage", 
              y="hwy",
              x="cty",
              title="Counts Plot") +
  theme(legend.position = "none" )
```

## 8월 평균계산

```r
install.packages('ggmap')
install.packages("readr")
library(ggmap)
library(readr)
```

```r
subway.data = read_csv('data/SUBWAY_flow2.csv')
```

- apply함수를 사용하여 합계필드(row.mean)를 만듭니다.

```r
row.mean = apply(subway.data[,-c(1:4)], 1, mean)
```

```r
row.mean %>%
  kable("html") %>%
  as.character() %>%
  display_html()
```

- cbind를 이용하여 원래 데이터에 row.mean필드를 더해 새로운 데이터프레임(subway.mean)을 만듭니다.

```r
subway.mean = cbind(subway.data , row.mean )
```

```r
subway.mean %>%
  kable("html") %>%
  as.character() %>%
  display_html()
```

- 위에 생성한 서울 지도 상에 row.neam필드를 점의 크기로 표현합니다. 점의 색상은 blue로 하고 투명도를 50%로 합니다.제목은 8월 평균 통행량, 부제는 학생명으로 합니다.

```r
register_google(key='')
```

```r
map.seoul = get_googlemap('seoul',
                          maptype = 'roadmap',
                          zoom = 11)
```

```r
ggmap(map.seoul) + geom_point(data = subway.mean,
                              aes(x = lon, y = lat, size = row.mean), color = 'blue', alpha = 0.5 ) + 
  labs(subtitle="XXX", 
       title="8월 평균 통행량")
```

## 2호선 지하철역의 특정일 날 통행량 시각화하기

- subset명령어를 이용하여 2호선인 데이터만 골라 새로운 데이터프레임(subway.line2)를 만듭니다.

```r
subway.line2 = subset(subway.data , line == '2호선')
```

```r
subway.line2 %>%
  kable("html") %>%
  as.character() %>%
  display_html()
```

- 위에 생성한 서울 지도 상에 8월 15일 필드(20200815)를 점의 크기로 표현합니다. 점의 색상은 dark green로 하고 투명도를 50%로 합니다.제목은 2호선 20200815 통행량, 부제는 학생명으로 합니다.

```r
ggmap(map.seoul) + geom_point(data = subway.line2,
                              aes(x = lon, y = lat, size = `20200815`), color = 'dark green', alpha = 0.5 ) + 
  labs(subtitle="XXX", 
       title="2호선 20200815 통행량")
```
